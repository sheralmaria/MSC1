```python
import pandas as pd
auto =pd.read_csv('Automobile_data.csv')
print(auto.head())
```

```python
import matplotlib.pyplot as plt
# create a figure and axis
fig, ax = plt.subplots()
ax.scatter(auto['horsepower'],auto['price'])
ax.set_title('Price to Performance Ratio')
ax.set_xlabel('Horsepower')
ax.set_ylabel('Price')
```

```python
import matplotlib.pyplot as plt
#create a figure and axis
fig, ax = plt.subplots()
ax.plot(auto['horsepower'],auto['price'])
plt.xticks(auto['horsepower'])
plt.yticks(auto['price'])
```

```python
#create figure and axis
fig, ax = plt.subplots()
#plot histogram
ax.hist(auto['horsepower'])
# set title and labels
ax.set_title('Automobile')
ax.set_xlabel('Horsepower')
ax.set_ylabel('Frequency')
```
